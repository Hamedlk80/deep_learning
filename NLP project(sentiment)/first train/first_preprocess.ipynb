```python
import pandas as pd
from hazm import Normalizer, word_tokenize

df = pd.read_csv("/content/Snappfood - Sentiment Analysis.csv", sep="\t", encoding="utf-8")

if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)

normalizer = Normalizer()

def preprocess(text):
    if pd.isna(text):
        return ""
    text = str(text)
    norm = normalizer.normalize(text)
    tokens = word_tokenize(norm)
    return " ".join(tokens)

df['processed'] = df['comment'].apply(preprocess)

df.to_csv('/content/snappfood_processed.csv', index=False)
df.head()
```

```python
import pandas as pd
from hazm import Normalizer, word_tokenize

digikala_df = pd.read_excel('/content/comment.xlsx')

print(digikala_df.columns)

digikala_df = digikala_df.dropna(subset=['comment'])

normalizer = Normalizer()

def preprocess(text):
    text = str(text)
    text = normalizer.normalize(text)
    tokens = word_tokenize(text)
    return " ".join(tokens)

digikala_df['processed'] = digikala_df['comment'].apply(preprocess)

digikala_df.to_excel('/content/digikala_processed.xlsx', index=False)
digikala_df.head()
```
